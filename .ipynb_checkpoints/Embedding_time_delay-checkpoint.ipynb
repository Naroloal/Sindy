```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

%matplotlib notebook
```

```python
'''Extraccion de datos, utilizamos unicamente pacientes 30,31,32 porque son los datos mas confiables'''
data = '/home/lorenzo/Principal/Noise-Neuron discriminator/Datos/PCCIBNMZc'
data = pd.read_pickle(data)

Neurons = data[data.bNoise == 0]
Neurons.reset_index(inplace = True)
```

```python
'''Ejemplo de media neuronal'''
neuron = Neurons.loc[32]
plt.plot(neuron.Mean)
```

```python
signal = neuron.Mean
```

```python
'''Funcion embedding'''
def signal_embedding(signal,d = 2,tau =1):
    '''Se le pasa la señal, las dimensiones y el tau. Devuelve el embedding como una serie de puntos '''
    embedding = signal[np.arange(d)*(tau+1) + np.arange(np.max(signal.shape[0] - (d-1)*(tau+1), 0)).reshape(-1,1)]
    return embedding.T
```

```python
'''Probamos con una unica senial'''
embedding = signal_embedding(signal,3,1)
```

```python
plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
x,y,z = embedding
ax.plot(x, y, z, label='parametric curve')
ax.legend()
plt.show()
```

```python
'''Ahora con todas las neuronas en Neurons, Primero en 3D'''
plt.rcParams['legend.fontsize'] = 10
Neurons['embedding'] = Neurons.Mean.apply(signal_embedding,args=(3,1))
```

```python
df = pd.DataFrame.from_records(Neurons.embedding.to_list(),columns= ['x','y','z'])
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
for i,neuron in df.iterrows():
    ax.plot(neuron.x,neuron.y,neuron.z)
```

```python
'''Ahora en 2D'''
plt.rcParams['legend.fontsize'] = 10

Neurons['embedding'] = Neurons.Mean.apply(signal_embedding,args=(2,1))
```

```python
df = pd.DataFrame.from_records(Neurons.embedding.to_list(),columns= ['x','y'])
```

```python
fig = plt.figure()
ax = fig.gca()
for i,neuron in df.iterrows():
    ax.plot(neuron.x,neuron.y)
```

```python
'''Clasificacion usando el ojimetro de la actividad neuronal'''
g0 = [2]
g1 = [24,29,40]
g2 = [23,16,24,25,26,15,27]
g3 = [28,29,30,14,31,32,33]
g4 = [34,13,35,36,37,38]
gs = np.array([g0,g1,g2,g3,g4])
Neurons['embedding'] = Neurons.Mean.apply(signal_embedding,args=(3,1))
dfs = {i:pd.DataFrame.from_records(Neurons.embedding.loc[g].to_list(),columns= ['x','y','z']) for i,g in enumerate(gs)}
```
